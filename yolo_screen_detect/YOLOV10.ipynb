```python
import os
from PIL import Image
from ultralytics import YOLOv10

# Khởi tạo mô hình với đường dẫn model_path
model_path = 'runs/detect/train12/weights/best.pt'
model = YOLOv10(model_path)

# Đường dẫn tới thư mục chứa ảnh đầu vào và thư mục lưu ảnh kết quả
input_folder = 'duc_data'
output_folder = 'processed_images'
os.makedirs(output_folder, exist_ok=True)

# Kích thước đầu ra cho ảnh
output_size = (960, 540)

# Duyệt qua từng ảnh trong thư mục đầu vào
for image_name in os.listdir(input_folder):
    image_path = os.path.join(input_folder, image_name)
    
    # Kiểm tra xem file có phải là ảnh hay không
    if not image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
        continue

    # Dự đoán và lưu kết quả
    results = model(image_path, conf=0.25)
    
    # Duyệt qua từng kết quả trong danh sách `results`
    for result in results:
        # Kiểm tra nếu kết quả có bounding boxes
        if hasattr(result, 'boxes') and result.boxes is not None:
            # Duyệt qua từng vùng phát hiện (bounding box)
            for idx, box in enumerate(result.boxes.xyxy):
                # Lấy bounding box
                x1, y1, x2, y2 = box[:4].cpu().numpy()
                bbox = (x1, y1, x2, y2)
                
                # Cắt vùng chứa điện thoại
                cropped_image = Image.open(image_path).crop(bbox)
                
                # Kiểm tra và xoay ảnh nếu là ảnh dọc
                if cropped_image.height > cropped_image.width:
                    cropped_image = cropped_image.rotate(90, expand=True)
                
                # Resize về kích thước 540x960
                resized_image = cropped_image.resize(output_size)
                
                # Chuyển đổi sang chế độ RGB nếu cần thiết
                if resized_image.mode != "RGB":
                    resized_image = resized_image.convert("RGB")
                
                # Lưu ảnh đã xử lý vào thư mục kết quả
                output_image_path = os.path.join(output_folder, f"{os.path.splitext(image_name)[0]}_cropped_{idx}.jpg")
                resized_image.save(output_image_path)
```
